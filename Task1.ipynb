# Checking unique value for each column

# Checking for empty data or missing value

```python
House_data.isnull().sum()
```

# Dividing columns into two parts Numeric and Categoric

```python
numerical_cols = House_data.select_dtypes(include=['int64', 'float64']).columns
categorical_cols = House_data.select_dtypes(include=['object', 'category']).columns
```

```python
numerical_cols
```

```python
categorical_cols
```

# Handelling Missing value

```python
from sklearn.impute import SimpleImputer

imputer_num = SimpleImputer(strategy='mean')
House_data[numerical_cols] = imputer_num.fit_transform(House_data[numerical_cols])
```

```python
numerical_cols.isnull().sum()
```

```python
# Handle missing values in categorical columns by dropping rows with missing values
# House_data = House_data.dropna(subset=categorical_cols)
House_data = House_data.fillna('None')
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoders = {}
for column in categorical_cols:
    label_encoders[column] = LabelEncoder()
    House_data[column] = label_encoders[column].fit_transform(House_data[column])
```

```python
House_data[categorical_cols].info()
```

```python
House_data[categorical_cols].head()
```

# Importing the Model

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score


model = LinearRegression()
```

# Dividing Data into Values and Labels

```python
x = House_data.drop('SalePrice', axis=1)
y = House_data['SalePrice']
```

```python
model.fit(x, y)
```

```python
y_pred = model.predict(x)
y_pred
```

```python
y
```

```python
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

# Calculate Mean Squared Error (MSE)
mse = mean_squared_error(y, y_pred)

# Calculate Root Mean Squared Error (RMSE)
rmse = np.sqrt(mse)

# Calculate R-squared
r2 = r2_score(y, y_pred)
mse , rmse  , r2
```
